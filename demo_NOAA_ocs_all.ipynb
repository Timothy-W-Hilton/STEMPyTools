This is a short demo which uses the noaa_ocs module and its
noaa_ocs.NOAA_OCS class to parse every NOAA OCS observation file from
a specified directory into one large <a
href="http://pandas.pydata.org/">pandas</a> <a
href="http://pandas.pydata.org/pandas-docs/stable/dsintro.html#dataframe">DataFrame</a>,
assign a STEM grid cell (X, Y, and Z) and timestep to every
observation, and write a STEM input.dat file from the NOAA data.

```python
#configure plotting 
#show figures in the notebook, not in a separate window.
%matplotlib inline
import matplotlib
#produce figures at 120 dpi, not the default 72
matplotlib.rcParams['savefig.dpi'] = 120
```

```python
#library modules
import os, os.path
import sys
import glob
import pandas as pd
import datetime
import matplotlib.pyplot as plt

#for writing a STEM input.dat file from NOAA observations
import inputdattools as idt
#for parsing STEM grid latitude, longitude, and topography
import STEM_parsers
# for parsing and plotting NOAA [OCS] observations
import noaa_ocs
```

Parse all of the OCS files in NOAA_95244993 to noaa_ocs.NOAA_OCS objects.

```python
noaa_dir = os.path.join(os.getenv('HOME'), 'work', 'Data', 'NOAA_95244993')
all_files = glob.glob(os.path.join(noaa_dir, '*ocs*.txt'))
data_list = [noaa_ocs.NOAA_OCS.parse_file(f) for f in all_files]
```

Now combine all of the DataFrames into one large DataFrame, and print
the first five rows.

```python
data = noaa_ocs.NOAA_OCS(
    obs=pd.concat([this_data.obs for this_data in data_list]))
default_cols_to_print = pd.get_option('display.max_columns')
pd.set_option('display.max_columns', len(data.obs.columns))
data.obs.head()
```

```python
pd.set_option('display.max_columns', default_cols_to_print)
```

a few observations list (longitude, latitude) of -999.999.  Remove those.

```python
data.obs = data.obs.iloc[data.obs['sample_longitude'].values > -998]
```

Next, assign STEM timesteps, STEM grid x and y, and STEM grid z to
each observation in the combined data frame.  Let's assume a STEM run
start time of 1 Jan 2008 00:00:00.

```python
STEM_t0 = datetime.datetime(2008, 1, 1)
STEM_grid_file = '/Users/tim/work/Data/STEM/input/TOPO-124x124.nc'
STEM_zhgt_file = '/Users/tim/work/Data/STEM/input/wrfheight-124x124-22levs.nc'
lon_stem, lat_stem, elev_stem = STEM_parsers.parse_STEM_coordinates(STEM_grid_file)
data.get_stem_t(STEM_t0)
data.get_stem_xy(lon_stem, lat_stem)
data.get_stem_z(STEM_grid_file, STEM_zhgt_file)
```

The above added the columns 'stemt', 'x_stem', 'y_stem', 'lon_stem',
'lat_stem', and 'z_stem' to the observations DataFrame:

```python
data.obs.iloc[:, -10:].head()
```

Show the data on a map:

```python
data.plot_locations()
plt.title('all NOAA [OCS] observation locations')
plt.show()
```

Many of the obsevations are well outside of the STEM grid.  Draw a plot that connects each observation to the nearest STEM grid cell:

```python
data.plot_stem_obs_matches()
plt.title('all NOAA [OCS] observation locations paired with STEM grid cells')
plt.show()
```

This looks good!  All of the the Pacific Ocean and Alaska observation
are connected to the nearest STEM grid cell.  There are no visible
lines on the rest of the map, meaning tat every observation was
matched to a nearby STEM cell.<BR><BR>

Last, write a STEM input.dat file from the parsed NOAA [OCS]
observations.  Let's assume a STEM run for 1 Jan 2008 through 31 Dec
2009.

```python
SECS_PER_HOUR = 60 * 60
STEM_end_tstep = (datetime.datetime(2009, 12, 31, 23, 59, 59) -
		  datetime.datetime(2008, 1, 1)).total_seconds() / SECS_PER_HOUR
idx = (data.obs.stemt.values >= 0) & (data.obs.stemt.values <= STEM_end_tstep)
data_in_window = data.obs.iloc[idx, :]
inputdat_maker = idt.STEMInputDat(hr=data_in_window.stemt.values,
				  x=data_in_window.x_stem.values,
				  y=data_in_window.y_stem.values,
				  z=data_in_window.z_stem.values,
				  obs=data_in_window.analysis_value.values,
				  hdr_note='##t, x, y, z, NOAA COS (ppbv)')
print inputdat_maker.write_file(fname='example_input.dat')
```

Print the first 10 lines of the input.dat file to verify it worked:

```python
f = open('example_input.dat', 'r')
for i in range(10):
    print(f.readline().strip())
f.close()
```
