## demo script: regridding CASA-GFED GEE to the STEM Lambert Conformal grid using the idw_kdtree_interp module.

This notebook tests the idw_kdtree_interp module's performance in
interpolating CASA-GFED GEE values from the CASA-GFED grid to the STEM
Lambert Conformal grid by producing a number of diagnostic plots.  By
way of these diagnostics it also demonstrates how to use the module
for interpolation.

#### First, set up some framework.

```python
#draw plots right in the notebook
%matplotlib inline
#produce figures at 120 dpi, not the default 72
import matplotlib
matplotlib.rcParams['savefig.dpi'] = 120

import os, os.path
import sys
import numpy as np
import netCDF4
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import mpl_toolkits.basemap as basemap


#get a nice qualitative colormap (details at http://colorbrewer2.org/)
import brewer2mpl
cmapdark2 = brewer2mpl.get_map('Dark2', 'Qualitative', number=8).mpl_colors

#nearest neighbor interpolation module
import idwnn_kdtree_interp as idw_kd
#for parsing STEM grid lat, lon
import STEM_parsers as sp
#for drawing the GEE on a map
import na_map
```

Define a function to show 2-D arrays on a map.  Using pcolormesh (as
opposed to contour) shows the data in the array without any
interpolation from the plotting routine.

```python
def plot_mesh(lon1, lat1, data1, title1, vmin=None, vmax=None,
	      cmap=cm.get_cmap('Greens_r')):
    fig = plt.figure(figsize=(10,20))

    m1 = na_map.NAMapFigure(cb_axis=True, t_str=title1)
    img = m1.map.pcolormesh(lon1,
		      lat1,
		      data1,
		      cmap=cmap,
		      latlon=True,
		      vmin=vmin,
		      vmax=vmax)
    plt.colorbar(img, cax=m1.ax_cmap)
    return(fig, m1)
    # return(fig, m1, m2)
```

```python
# get CASA lat, lon, GEE
nc = netCDF4.Dataset(os.path.join(os.getenv('HOME'),
                                  'work/Data/CASA_GFED/CASA_GEE_154hrs_July2008.nc'))
lon_casa = nc.variables['lon'][:]
lat_casa = nc.variables['lat'][:]
gee = nc.variables['GEE'][:].squeeze()
nc.close()
# get 2D arrays of latitudes, longitudes
lat_casa, lon_casa = np.meshgrid(lat_casa, lon_casa, indexing='ij')

#get STEM lon, lat
fname_STEM_coords = '/Users/tim/work/Data/STEM/input/TOPO-124x124.nc'
lon_stem, lat_stem, topo_stem = sp.parse_STEM_coordinates(fname_STEM_coords)
```

#### test 1: data in two dimensions (that is, say, a single timestep)

This takes a 181 by 288 array of CASA-GFED GEE data and interpolates
it onto the 124 by 124 STEM grid, using the CASA-GFED grid's 9 nearest
neighbors (the default) to each STEM grid cell to determine the
values.

```python
#interpolate CASA GEE onto STEM grid
gee_stem_grid = idw_kd.spherical_idw_kdtree_interp(lon_casa,
						   lat_casa,
						   lon_stem,
						   lat_stem,
						   gee[0,...])
```

```python
fig1, m1 = plot_mesh(lon_stem, lat_stem, gee_stem_grid,
		    'CASA-GFED GEE interpolated to STEM grid',
		    vmin=gee[0, ...].min(), vmax=gee[0, ...].max())
fig2, m2 = plot_mesh(lon_casa, lat_casa, gee[0, ...],
		    'CASA-GFED GEE on CASA-GFED grid',
		    vmin=gee[0, ...].min(), vmax=gee[0, ...].max())
```

#### test 2: data in three dimensions (that is, say, many single timesteps)

Plot a few timesteps of GEE data to make sure the daily cycle looks
reasonable.

```python
fig3, m13 = plot_mesh(lon_casa, lat_casa, gee[0,...],
		    'CASA-GFED GEE hour 0',
		    vmin=gee.min(), vmax=gee.max())
fig4, m4 = plot_mesh(lon_casa, lat_casa, gee[3,...],
		    'CASA-GFED GEE hour 9',
		    vmin=gee.min(), vmax=gee.max())
fig5, m5 = plot_mesh(lon_casa, lat_casa, gee[7,...],
		    'CASA-GFED GEE hour 21',
		    vmin=gee.min(), vmax=gee.max())
```

Now interpolate the three-dimensional (time, longitude, latitude)
CASA-GFED GEE onto the STEM grid using two nearest neighbors for each
interpolation, and again using nine nearest neighbors.

```python
gee_i1 = idw_kd.spherical_idw_kdtree_interp(lon_casa, lat_casa,
					    lon_stem, lat_stem,
					    gee,
					    n_nbr=1)
gee_i2 = idw_kd.spherical_idw_kdtree_interp(lon_casa, lat_casa,
					    lon_stem, lat_stem,
					    gee,
					    n_nbr=2)
gee_i9 = idw_kd.spherical_idw_kdtree_interp(lon_casa, lat_casa,
					    lon_stem, lat_stem,
					    gee,
					    n_nbr=9)
```

...and plot a single point's (95$^{\circ}$W, 42$^{\circ}$) 154-hour
timeseries for both raw CASA-GFED GEE as well as the interpolated GEE.

```python
# translate (95W, 42N), all STEM grid coordinates, and all CASA-GFED
# grid coordinates to Cartesian X, Y, Z.
x, y, z = idw_kd.lon_lat_to_cartesian(-95, 42)
x_stem, y_stem, z_stem = idw_kd.lon_lat_to_cartesian(lon_stem, lat_stem)
x_casa, y_casa, z_casa = idw_kd.lon_lat_to_cartesian(lon_casa, lat_casa)
# find the indices of the grid cell nearest to (95W, 42N) in STEM and
# CASA grids
d_stem, idx_stem_flat = idw_kd.find_nn_3D_kdtree(x_stem, y_stem, z_stem,
						 x, y, z,
						 n_nbr=1)
d_casa, idx_casa_flat = idw_kd.find_nn_3D_kdtree(x_casa, y_casa, z_casa,
						 x, y, z,
						 n_nbr=1)
#inds_stem and inds_casa are flat indices; get the 2-dimensional indices
idx_stem = np.unravel_index(idx_stem_flat, lon_stem.shape)
idx_casa = np.unravel_index(idx_casa_flat, lon_casa.shape)
# now plot the timeseries
fig, ax = plt.subplots()
VALS_PER_DAY = 8  #CASA GEE is 3-hourly
t_idx = np.arange(VALS_PER_DAY * 5)
line_raw = ax.plot(t_idx,
                   gee[t_idx, idx_casa[0], idx_casa[1]],
                   color='black',
                   label='CASA GEE (raw)')
line_interp2 = ax.plot(t_idx,
                      gee_i2[t_idx, idx_stem[0], idx_stem[1]],
                      color=cmapdark2[0],
                      linestyle='dashed',
                      label='CASA GEE (interp., 2 nbrs)')
line_interp2 = ax.plot(t_idx,
                      gee_i9[t_idx, idx_stem[0], idx_stem[1]],
                      color=cmapdark2[1],
                      linestyle='dashed',
                      label='CASA GEE (interp., 9 nbrs)')
plt.title('CASA pseudo-GEE w/ daily cycle, (95$^{\circ}$W, 42$^{\circ}$N)')
plt.xlabel('hours from t$_0$')
plt.ylabel('GEE')
fontP = FontProperties()
fontP.set_size('small')
plt.legend(loc='best', prop=fontP)
```

It looks like interpolating with two nearest neighbors captures the
daily GEE peaks better than 9 nearest neighbors, at least at
95$^{\circ}$W, 42$^{\circ}$N.  Using fewer nearest neighbors for
interpolation also does less smoothing of the CASA GEE, resulting in a
more truthful representation of the CASA spatial resolution.

```python
t_idx = 8
fig3, m13 = plot_mesh(lon_casa, lat_casa, gee[t_idx,...],
		    'CASA-GFED GEE, hour 24, on native grid',
		    vmin=gee.min(), vmax=gee.max())
fig3, m13 = plot_mesh(lon_stem, lat_stem, gee_i1[t_idx,...],
		    'CASA-GFED GEE, hour 24, on STEM grid (1 nbr)',
		    vmin=gee.min(), vmax=gee.max())
fig4, m4 = plot_mesh(lon_stem, lat_stem, gee_i2[t_idx,...],
		    'CASA-GFED GEE, hour 24, on STEM grid (2 nbrs)',
		    vmin=gee.min(), vmax=gee.max())
fig5, m5 = plot_mesh(lon_stem, lat_stem, gee_i9[t_idx,...],
		    'CASA-GFED GEE, hour 24, on STEM grid (9 nbrs)',
		    vmin=gee.min(), vmax=gee.max())
```

#### How well does nearest-neighbor interpolation preserve total mass?

CASA's GEE units are kg C m$^{-2}$ s$^{-1}$.  I'd like to compare
North America continental integrated CO$_2$ for 1, 2, and 9 (and
other?) nearest neighbors interpolations. However, getting a true
integrated flux for each grid cell is somewhat more challenging than I
initially realized, because integrating kg C m$^{-2}$ s$^{-1}$ to
something like kg C day$^{-1}$ requires the area of each STEM grid
cell in m$^2$.  This is non-trivial because the Lambert conformal
projection does not preserve area, so each STEM grid cell has a
different area.  I need to think about this a little more.  I'd also
want the integrated CASA GEE for comparison, and that requires similar
calculations for the CASA grid.

```python
#convert kgC m-2 s-1 to kgC m-2
SECS_PER_HOUR = 60 * 60
HOURS_PER_TSTEP = 3
n_tsteps = gee_i9.shape[0]
f = SECS_PER_HOUR * HOURS_PER_TSTEP   # conversion factor, kg C m-2 s-1 to kg C m-2
err = np.sum(gee_i9 - gee_i1, axis=0) * f
fig, m = plot_mesh(lon_stem, lat_stem, err,
		   'total interpolation error, 9 nbrs - 1 nbr',
		    vmin=-np.abs(err).max(),
		    vmax=np.abs(err).max(),
            cmap=cm.get_cmap('PuOr'))
m.ax_cmap.set_title('kgC m$^{-2}$')
```

Most of the errors are small.  This looks good for inverse distance weighting.  I'm going to try integrating the continental fluxes assuming every STEM grid cell is 60 km by 60 km.  This might not be accurate, but probably isn't drastically off.

```python
def kg2Pg(kg):
    """
    convert kilograms (1e3 g) to petagrams (1e15 g).
    """
    return(kg * 1e-12)

#convert kgC m-2 s-1 to kgC
m2_per_cell = 60000 * 60000
f = SECS_PER_HOUR * HOURS_PER_TSTEP * m2_per_cell #conversion factor
kgC_i9 = np.sum(gee_i9 * f)
kgC_i1 = np.sum(gee_i1 * f) 
print "9 neighbors total GPP (Pg C): ", kg2Pg(kgC_i9)
print "1 neighbor total GPP (Pg C):", kg2Pg(kgC_i1)
print "percent change, 1 NN to 9 NN: ", ((kgC_i1 - kgC_i9) / kgC_i1) * 100
```

<a href="http://dx.doi.org/10.1016/j.ecolmodel.2012.02.004">Huntzinger et al (2012)</a> compared a range of land surface models and found GPPs of roughly 3 Pg C month$^{-2}$ for North America during the growing season.  This CASA-GFED GEE subset covers 19 days so the total GEE is definitely in the right ballpark.  This is encouraging:  the inverse distance weighted nearest neighbor interpolation seems fairly insensitive to the number of nearest neighbors used, and produces results consistent with published models.
