## demo script: regridding CASA-GFED GEE to the STEM Lambert Conformal grid using the idw_kdtree_interp module.

This notebook demonstrates using the idw_kdtree_interp module to
interpolate CASA-GFED GEE values from the CASA-GFED grid to the STEM
Lambert Conformal grid.

#### First, set up some framework.

```python
#draw plots right in the notebook
%matplotlib inline
#produce figures at 120 dpi, not the default 72
import matplotlib
matplotlib.rcParams['savefig.dpi'] = 120

import os, os.path
import sys
import numpy as np
import netCDF4
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import mpl_toolkits.basemap as basemap

import idwnn_kdtree_interp as idw_kd

#for parsing STEM grid lat, lon
sys.path.append('/Users/tim/work/Code/STEM_vis_py/')
import STEM_parsers as sp
#for drawing the GEE on a map
import na_map
```

Define a function to show 2-D arrays on a map.  Using pcolormesh (as
opposed to contour) shows the data in the array without any
interpolation from the plotting routine.

```python
def plot_mesh(lon1, lat1, data1, title1, vmin=None, vmax=None):
    fig = plt.figure(figsize=(10,20))

    m1 = na_map.NAMapFigure(cb_axis=True, t_str=title1)
    img = m1.map.pcolormesh(lon1,
		      lat1,
		      data1,
		      cmap=cm.get_cmap('Greens_r'),
		      latlon=True,
		      vmin=vmin,
		      vmax=vmax)
    plt.colorbar(img, cax=m1.ax_cmap)
    return(fig, m1)
    # return(fig, m1, m2)
```

```python
# get CASA lat, lon, GEE
nc = netCDF4.Dataset(os.path.join(os.getenv('HOME'),
                                  'work/Data/CASA_GFED/CASA_GEE_slice_July2008.nc'))
lon_casa = nc.variables['lon'][:]
lat_casa = nc.variables['lat'][:]
gee = nc.variables['GEE'][:].squeeze()
nc.close()
# get 2D arrays of latitudes, longitudes
lat_casa, lon_casa = np.meshgrid(lat_casa, lon_casa, indexing='ij')

#get STEM lon, lat
fname_STEM_coords = '/Users/tim/work/Data/STEM/input/TOPO-124x124.nc'
lon_stem, lat_stem, topo_stem = sp.parse_STEM_coordinates(fname_STEM_coords)
```

#### test 1: data in two dimensions (that is, say, a single timestep)

This takes a 181 by 288 array of CASA-GFED GEE data and interpolates
it onto the 124 by 124 STEM grid, using the CASA-GFED grid's 9 nearest
neighbors (the default) to each STEM grid cell to determine the
values.

```python
#interpolate CASA GEE onto STEM grid
gee_stem_grid = idw_kd.spherical_idw_kdtree_interp(lon_casa,
						   lat_casa,
						   lon_stem,
						   lat_stem,
						   gee)
```

```python
fig1, m1 = plot_mesh(lon_stem, lat_stem, gee_stem_grid,
		    'CASA-GFED GEE interpolated to STEM grid',
		    vmin=gee.min(), vmax=gee.max())
fig2, m2 = plot_mesh(lon_casa, lat_casa, gee,
		    'CASA-GFED GEE on CASA-GFED grid',
		    vmin=gee.min(), vmax=gee.max())
```

#### test 2: data in three dimensions (that is, say, many single timesteps)

First, create two days worth of hourly GEE pseudo-data by adding a
daily cycle to the single-timestep CASA-GFED GEE from above.

```python
#tile the GEE matrix out to 48 timesteps and impose a sinusoidal cycle
#with values between 0 and 1.0 on it to simulate hourly GEE for two
#days
n_hrs = 48
gee3d = np.tile(gee, (n_hrs, 1, 1))
f = np.array(np.linspace(start=np.pi/2,
			 stop=(9*np.pi)/2,
			 num=n_hrs)).reshape((-1,1,1))
gee3d = gee * ((np.sin(f) + 1.0) / 2.0)
```

Plot a few timesteps of GEE data to make sure the pseudo daily cycle
looks reasonable.

```python
fig3, m13 = plot_mesh(lon_casa, lat_casa, gee3d[0,...],
		    'CASA-GFED GEE hour 0',
		    vmin=gee.min(), vmax=gee.max())
fig4, m4 = plot_mesh(lon_casa, lat_casa, gee3d[8,...],
		    'CASA-GFED GEE hour 12',
		    vmin=gee.min(), vmax=gee.max())
fig5, m5 = plot_mesh(lon_casa, lat_casa, gee3d[20,...],
		    'CASA-GFED GEE hour 20',
		    vmin=gee.min(), vmax=gee.max())
```

And plot a single point's "two day" timeseries.

```python
np.unravel_index(np.argmin(gee[0,...]), gee.shape)
plt.plot(np.arange(gee3d.shape[0]), gee3d[:,131, 69])
plt.title('CASA GEE with simulated daily cycle, x=131, y=69')
plt.xlabel('hours from t$_0$')
plt.ylabel('GEE')
```

Looks like the "pseudo daily cycle" was successful.  Now try to
interpolate the 48 by 181 by 288 GEE array to make sure the
interpolation routine works with three dimensional input.  Use the
default of nine nearest neighbors for the interpolation.

```python
gee3di = gee_stem_grid = idw_kd.spherical_idw_kdtree_interp(lon_casa,
							    lat_casa,
							    lon_stem,
							    lat_stem,
							    gee3d)
```

```python
fig3, m13 = plot_mesh(lon_casa, lat_casa, gee3d[20,...],
		    'CASA-GFED GEE, hour 20, on native grid',
		    vmin=gee.min(), vmax=gee.max())
fig4, m4 = plot_mesh(lon_stem, lat_stem, gee3di[20,...],
		    'CASA-GFED GEE, hour 20, interpolated to STEM grid',
		    vmin=gee.min(), vmax=gee.max())
```

Using fewer nearest neighbors for interpolation will smooth the CASA
GEE less and therefore produce a more faithful representation of the
CASA GEE on its native grid.  Try the interpolation using 1 nearest
neighbor instead of 9.

```python
n_nbr = 1
gee3di_nbrN = idw_kd.spherical_idw_kdtree_interp(lon_casa,
						 lat_casa,
						 lon_stem,
						 lat_stem,
						 gee3d,
						 n_nbr=n_nbr)
fig3, m13 = plot_mesh(lon_casa, lat_casa, gee3d[20,...],
		    'CASA-GFED GEE, hour 20',
		    vmin=gee.min(), vmax=gee.max())
fig4, m4 = plot_mesh(lon_stem, lat_stem, gee3di_nbrN[20,...],
		    'CASA-GFED GEE, hour 20, on STEM grid, n_nbr={}'.format(n_nbr),
		    vmin=gee.min(), vmax=gee.max())
```
