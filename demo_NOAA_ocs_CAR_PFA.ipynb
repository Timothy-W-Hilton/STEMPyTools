#### Parsing NOAA OCS observations and finding the nearest grid cell

This is a short demonstration of the noaa_ocs module and
noaa_ocs.NOAA_OCS class.  The demo:<BR>
(1) parses two NOAA airborne observation files for sites PFA and CAR.<BR>
(2) assigns each observation to a horizontal STEM grid cell (X, Y)<BR>
(3) assigns each obseration to a STEM timestemp assuming a (arbitrary)
STEM start time of 1 June 2008 00:00:00.<BR>
(4) assigns each observation to a vertical STEM grid cell Z.  <BR>

```python
%matplotlib inline
```

```python
#libraries
import os, os.path
import numpy as np
from mpl_toolkits import basemap
import matplotlib
import matplotlib.pyplot as plt
from datetime import datetime
from scipy.io import netcdf

#for parsing latitude, longitude of STEM grid
import STEM_parsers
#tools for parsing, plotting NOAA OCS data
import noaa_ocs
```

```python
matplotlib.rcParams['savefig.dpi'] = 120  #produce figures at 120 dpi, not the default 72
```

Parse two flights worth of NOAA OCS data from
ocs_car_aircraft-pfp_1_hats_event.txt and
ocs_pfa_aircraft-pfp_1_hats_event.txt (chosen more or less at random
from the files on ecampbell300):

```python
ocs_car = 'ocs_car_aircraft-pfp_1_hats_event.txt'
ocs_pfa = 'ocs_pfa_aircraft-pfp_1_hats_event.txt'
ocs_car = noaa_ocs.NOAA_OCS.parse_file(
    os.path.join('/Users', 'tim', 'work',
    'Data', 'NOAA_95244993', ocs_car))
ocs_pfa = noaa_ocs.NOAA_OCS.parse_file(
    os.path.join('/Users', 'tim', 'work',
    'Data', 'NOAA_95244993', ocs_pfa))
```

For each flight, assign each observation to the nearest STEM grid cell
center:

```python
STEM_grid_file = '/Users/tim/work/Data/STEM/input/TOPO-124x124.nc'
lon_stem, lat_stem, elev_stem = STEM_parsers.parse_STEM_coordinates(STEM_grid_file)
ocs_car.get_stem_xy(lon_stem, lat_stem)
ocs_pfa.get_stem_xy(lon_stem, lat_stem)
```

Plot the observation locations and corresponding STEM grid cell centers for the
two flights on a map:

```python
m_obs, han  = ocs_car.plot_locations()
m_obs.scatter(lon_stem[ocs_car.obs['x_stem'].values, ocs_car.obs['y_stem'].values],
	      lat_stem[ocs_car.obs['x_stem'].values, ocs_car.obs['y_stem'].values],
	      color=ocs_car.grid_color,
	      marker='^',
          label='STEM grid cell')
m_obs.scatter(ocs_pfa.obs.sample_longitude,
	      ocs_pfa.obs.sample_latitude,
	      color=ocs_car.obs_color,
	      marker='.',
          label='observation')
m_obs.scatter(lon_stem[ocs_pfa.obs['x_stem'].values, ocs_pfa.obs['y_stem'].values],
	      lat_stem[ocs_pfa.obs['x_stem'].values, ocs_pfa.obs['y_stem'].values],
	      color=ocs_pfa.grid_color,
	      marker='^')
plt.legend(numpoints=1,
           scatterpoints=1,
           loc='best')
plt.show()
```

It looks like the pfa flight, in Alaska, is outside of the STEM grid.

To verify that the grid cell assignment algorithm works, show each
observation from ocs_car_aircraft-pfp_1_hats_event.txt paired with its
assigned STEM grid cell center.

```python
m_matches = ocs_car.plot_stem_obs_matches()
plt.title('NOAA [OCS] observations, site: CAR')
plt.show()
```

Looks good!

Now, calculate STEM time (hours past STEM start time t$_0$).  Let's
assume a STEM t$_0$ of 1 June 2008.

```python
ax = plt.axes()
stem_t0 = datetime(2008, 6, 1)
ocs_car.get_stem_t(stem_t0)
plt.plot_date(ocs_car.obs.datet.astype(datetime), ocs_car.obs.stemt)
plt.ylabel('hours from {}'.format(stem_t0.strftime('%d %b %Y')))
plt.xlabel('observation date')
plt.title('NOAA [OCS] observations, site: CAR')
```

Assign STEM z levels (1 to 22) to each NOAA observation.  The z level
heights come from WRF pressure surfaces, which are themselves 124 by
124 by 22.  That is, the height (in meters above sea level) for a
given z level varies horizontally. This horizontal variation is why
the assigned z level does not increase monotonically with sample
alitude.  The horizontal variation also means that the Z level
assignment has to happen after the horizontal grid cell assignment.

```python
wrfheight_fname='/Users/tim/work/Data/STEM/input/wrfheight-124x124-22levs.nc'
ocs_car.get_stem_z(topo_fname=STEM_grid_file,
		   wrfheight_fname=wrfheight_fname)
plt.plot(ocs_car.obs.sample_altitude.values,
	 ocs_car.obs.z_stem.values,
	 'ok')
plt.xlabel('sample altitude (m above sea level)')
plt.ylabel('assigned STEM z level')
plt.title('NOAA [OCS] observations, site: CAR')
```
