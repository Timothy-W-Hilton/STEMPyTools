#### Parsing NOAA OCS observations and finding the nearest grid cell

```python
%matplotlib inline
```

```python
#libraries
import os, os.path
import numpy as np
from mpl_toolkits import basemap
import matplotlib
import matplotlib.pyplot as plt

#for parsing latitude, longitude of STEM grid
import STEM_parsers
#tools for parsing, plotting NOAA OCS data
import noaa_ocs
```

```python
matplotlib.rcParams['savefig.dpi'] = 120  #produce figures at 120 dpi, not the default 72
```

Parse two flights worth of NOAA OCS data from
ocs_car_aircraft-pfp_1_hats_event.txt and
ocs_pfa_aircraft-pfp_1_hats_event.txt (chosen more or less at random
from the files on ecampbell300):

```python
ocs_car = 'ocs_car_aircraft-pfp_1_hats_event.txt'
ocs_pfa = 'ocs_pfa_aircraft-pfp_1_hats_event.txt'
ocs_car = noaa_ocs.NOAA_OCS.parse_file(
    os.path.join('/Users', 'tim', 'work',
    'Data', 'NOAA_95244993', ocs_car))
ocs_pfa = noaa_ocs.NOAA_OCS.parse_file(
    os.path.join('/Users', 'tim', 'work',
    'Data', 'NOAA_95244993', ocs_pfa))
```

For each flight, assign each observation to the nearest STEM grid cell
center:

```python
STEM_grid_file = '/Users/tim/work/Data/STEM/input/TOPO-124x124.nc'
lon_stem, lat_stem, elev_stem = STEM_parsers.parse_STEM_coordinates(STEM_grid_file)
ocs_car.get_stem_xy(lon_stem, lat_stem)
ocs_pfa.get_stem_xy(lon_stem, lat_stem)
```

Plot the observation locations and corresponding STEM grid cell centers for the
two flights on a map:

```python
m_obs, han  = ocs_car.plot_locations()
m_obs.scatter(lon_stem[ocs_car.obs['x_stem'].values, ocs_car.obs['y_stem'].values],
	      lat_stem[ocs_car.obs['x_stem'].values, ocs_car.obs['y_stem'].values],
	      color=ocs_car.grid_color,
	      marker='^',
          label='STEM grid cell')
m_obs.scatter(ocs_pfa.obs.sample_longitude,
	      ocs_pfa.obs.sample_latitude,
	      color=ocs_car.obs_color,
	      marker='.',
          label='observation')
m_obs.scatter(lon_stem[ocs_pfa.obs['x_stem'].values, ocs_pfa.obs['y_stem'].values],
	      lat_stem[ocs_pfa.obs['x_stem'].values, ocs_pfa.obs['y_stem'].values],
	      color=ocs_pfa.grid_color,
	      marker='^')
plt.legend(numpoints=1,
           scatterpoints=1,
           loc='best')
plt.show()
```

It looks like the pfa flight, in Alaska, is outside of the STEM grid.
